## Exercise 1 - Understanding the problem and Data Collection

We want to predict if an individual has diabetes.

- Load the diabetes dataset and explore it
- Count positive and negative cases
- Split the data into train and test


```python

# TODO: load the dataset
# If running on Colab, upload the zip or csv then adjust the path
import pandas as pd


df= pd.read_csv("/content/diabetes_prediction_dataset.csv")

print(df.shape)
display(df.head())
print(df.dtypes)
print("Missing per column:")
display(df.isna().sum().sort_values(ascending=False))
```

```python
# Assume target column is named 'diabetes' with 0 or 1 values
assert 'diabetes' in df.columns, "Expected a 'diabetes' target column"
print(df['diabetes'].value_counts())
```

```python

# TODO: train test split
from sklearn.model_selection import train_test_split
X = df.drop(columns=['diabetes'])
y = df['diabetes']
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2, # 20% for testing
    random_state=42 # seed
)
print(X_train.shape, X_test.shape)
```

- Do we need to standardize
- If yes, apply StandardScaler



> TODO: In a short paragraph, justify Logistic Regression for this binary task. Mention linear decision boundary, calibrated probabilities, and interpretability. Explain why standardization helps for numerical stability and better conditioning.

Certainly! Here are the explanations for your exercise regarding Logistic Regression and Data Standardization in the context of the diabetes_prediction_dataset.csv.

1. Model Selection: Why Logistic Regression?
For a binary classification task (predicting whether a patient has diabetes or not), Logistic Regression is often the ideal starting point for the following reasons:

Linear Decision Boundary: It is highly efficient when the relationship between the features (like BMI, Age, Glucose) and the probability of having diabetes can be separated by a linear boundary.

Calibrated Probabilities: Unlike some models that only give a "yes/no" answer, Logistic Regression provides a probability score (e.g., 0.85 chance of diabetes). In medical contexts, this helps clinicians understand the level of risk rather than just a hard classification.

Interpretability: It is a "white-box" model. By looking at the model's coefficients, you can clearly see which factors (like HbA1c_level) have the strongest impact on the prediction.

2. Why Standardize the Data?
Standardization (specifically using StandardScaler) is crucial for this model because:

Numerical Stability: Logistic Regression uses optimization algorithms (like Gradient Descent). If one feature has a range of 0–1 (like heart_disease) and another has 0–200 (like blood_glucose_level), the algorithm may struggle to converge or take much longer to find the optimal solution.

Fair Feature Weighting: Without scaling, the model might "think" that features with larger numerical values are more important than those with smaller values, simply because of their scale.

Better Conditioning: It ensures the cost function is more spherical, allowing the solver to reach the minimum more efficiently.

```python
# TODO: build a preprocessing pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

cat_cols = X.select_dtypes(include=['object']).columns.tolist()
num_cols = X.select_dtypes(include=['int64','float64']).columns.tolist()

preprocess = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
])
print("Categorical:", cat_cols)
print("Numeric:", num_cols)

```

## Exercise 3 - Model training

```python

# TODO: train Logistic Regression
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

clf = Pipeline([
    ('preprocessor', preprocess),
    ('classifier', LogisticRegression(max_iter=1000))
])

# Fit the model to the training data
clf.fit(X_train, y_train)
```


## Exercise 4 - Evaluation metrics

- Plot accuracy and comment
- Plot confusion matrix and comment
- Plot precision, recall, F1 and comment


```python
# TODO: use the metrics functions properly to plot the scores
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

y_pred = clf.predict(X_test)

comparison = pd.DataFrame({
    'Actual': y_test.values[:10],
    'Predicted': y_pred[:10],
    'Match': ['MATCH' if a == p else 'NOT' for a, p in zip(y_test.values[:10], y_pred[:10])]
})

print()
print("First 10 predictions vs actual:")
display(comparison)

correct = (y_pred == y_test.values).sum()
total = len(y_test)

print()
print(f"Overall: {correct} correct out of {total} ({correct/total*100:.1f})%")
print(f"Predicted {y_pred.sum()} diabetic - actual {y_test.sum()} diabetic")


from sklearn.metrics import (accuracy_score, confusion_matrix,
                             ConfusionMatrixDisplay, classification_report,
                             precision_score, recall_score, f1_score)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)


print("Accuracy:", round(acc,4))
print("Precision:", round(prec,4))
print("Recall:", round(rec,4))
print("F1:", round(f1,4))

# Simple bar plot of metrics
import matplotlib.pyplot as plt
plt.figure()
plt.bar(['accuracy','precision','recall','f1'], [acc,prec,rec,f1])
plt.title('Metrics on test')
plt.ylabel('Score')
plt.show()

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title('Confusion matrix')
plt.show()
```

> TODO: comment on the balance between precision and recall.

Precision vs. Recall Balance
In this model, we see high Precision (0.86) but lower Recall (0.62):

High Precision: When the model predicts someone has diabetes, it is correct about 86% of the time. There are very few "false alarms" (167 False Positives).

Lower Recall: The model only manages to catch about 62% of the actual diabetes cases.

Medical Impact: In a healthcare setting, a low recall is often risky because it means we are missing nearly 40% of people who actually need treatment. To improve this, we might need to adjust the classification threshold to prioritize recall over precision.


## Exercise 5 - Visualizing the performance of our model

Visualize a 2D decision boundary with accuracy info. Use only two informative features for this plot to keep it 2D. Suggested pair: `HbA1c_level` and `blood_glucose_level` if present. Otherwise pick any two numeric features.


```python
# TODO: If these columns do not exist, change `feat_x` and `feat_y` below to two numeric features that exist in your data.
import numpy as np
import matplotlib.pyplot as plt

feat_x = 'HbA1c_level' if 'HbA1c_level' in X.columns else X.select_dtypes(include=['int64','float64']).columns[0]
feat_y = 'blood_glucose_level' if 'blood_glucose_level' in X.columns else X.select_dtypes(include=['int64','float64']).columns[1]

X2_train = X_train[[feat_x, feat_y]].copy()
X2_test = X_test[[feat_x, feat_y]].copy()

pipe2 = Pipeline([
    ('pre', ColumnTransformer([('num', StandardScaler(), [0,1])], remainder='drop')),
    ('lr', LogisticRegression(max_iter=1000, random_state=42))
])
pipe2.fit(X2_train.values, y_train)

# Mesh
x_min, x_max = X2_train[feat_x].min()-1, X2_train[feat_x].max()+1
y_min, y_max = X2_train[feat_y].min()-1, X2_train[feat_y].max()+1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
probs = pipe2.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1].reshape(xx.shape)

plt.figure(figsize=(6,5))
cs = plt.contour(xx, yy, probs, levels=[0.5])
plt.clabel(cs, inline=True, fmt={0.5:'P=0.5'})
plt.scatter(X2_test[feat_x], X2_test[feat_y], c=y_test, alpha=0.7)
plt.xlabel(feat_x); plt.ylabel(feat_y)
from sklearn.metrics import accuracy_score
acc2 = accuracy_score(y_test, pipe2.predict(X2_test.values))
plt.title(f'Decision boundary on 2 features - test accuracy {acc2:.3f}')
plt.show()
```


## Exercise 6 - ROC curve

Use the code template provided to plot the ROC curve for your model and compute AUC. You can reuse the fitted `clf` pipeline.

Template summary:
- Get predicted probabilities for the positive class
- Compute fpr and tpr with `roc_curve`
- Plot ROC and print AUC


```python

from sklearn import metrics
y_proba = clf.predict_proba(X_test)[:,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_proba)
auc = metrics.roc_auc_score(y_test, y_proba)

plt.figure()
plt.plot(fpr, tpr, label=f'AUC={auc:.3f}')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.title('ROC curve')
plt.show()
```

> TODO: interpret the ROC curve and AUC from results of execution of cell above.

Interpretation of the ROC Curve and AUC
AUC Score (0.962): An Area Under the Curve (AUC) of 0.962 is excellent. It indicates that there is a 96.2% chance that the model will be able to distinguish between a patient with diabetes and a healthy patient.

Curve Shape: The curve climbs very quickly toward the top-left corner. This means the model achieves a high True Positive Rate (Sensitivity) while maintaining a low False Positive Rate.

Performance vs. Random: The orange line is significantly above the dashed navy line (which represents a random classifier). This confirms the model is highly effective and far better than guessing.

Clinical Utility: In a medical context, this curve shows we have the flexibility to move the classification threshold. If we want to catch even more diabetes cases (increase Recall/Sensitivity), we can move along the curve, though this would slightly increase the number of false alarms.

Overall, the high AUC confirms that the Logistic Regression model is very robust for this dataset.
