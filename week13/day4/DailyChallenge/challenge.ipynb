```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#use this website : Github/topics
url = "https://github.com/topics"

#Write a Python script using the requests library to fetch the HTML content of the chosen website.
response = requests.get(url)

#Print the status code of the response to ensure the request was successful using .status_code, it should be 200
print(f"Status code: {response.status_code}")

#Print the first 100 characters of the HTML content to verify the response.
print(f"First 100 characters:\n{response.text[:100]}")

#Save the HTML content to a file named webpage.html. Ensure you handle the text encoding correctly.
with open('webpage.html', 'w', encoding='utf-8') as f:
    f.write(response.text)
```

```python
#Use BeautifulSoup to parse the saved HTML content from the file
with open('webpage.html', 'r', encoding='utf-8') as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# Identify and Extract Information
# We are looking for Topic Titles and their Descriptions

topic_title_tags = soup.select('p.f3')
topic_desc_tags = soup.select('p.f5')

topic_titles = []
for tag in topic_title_tags:
  topic_titles.append(tag.get_text())

topic_descs = []
for tag_descs in topic_desc_tags:
  topic_descs.append(tag_descs.get_text())

#Print the length and content of each extracted list to verify the extraction process.
print(f"\nExtracted {len(topic_titles)} titles.")
print(f"Extracted {len(topic_descs)} descriptions.")

for titles in topic_titles:
  print(titles)


for desc in topic_descs:
  print(desc)

#Create a Python dictionary to structure the extracted data, with keys representing the type of information (e.g., ‘title’ and ‘description’)
topics_dict = {
    'title': topic_titles,
    'description': topic_descs
}
```

```python
#Convert this dictionary into a pandas DataFrame.
df = pd.DataFrame(topics_dict)

# Print the DataFrame
print("\n--- Final DataFrame ---")
print(df.head())
```
