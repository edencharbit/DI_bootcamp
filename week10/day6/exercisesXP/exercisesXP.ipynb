```python
# Exercise 2

import numpy as np
from scipy import stats

# Sample Dataset
data = [12, 15, 13, 12, 18, 20, 22, 21]

# 1. Using the 'describe' function for a quick overview
description = stats.describe(data)

# 2. Calculating specific metrics
mean_val = np.mean(data)
median_val = np.median(data)
variance_val = stats.tvar(data)  # Sample variance
std_dev_val = stats.tstd(data)   # Sample standard deviation

print(f"Dataset: {data}")
print("-" * 30)
print(f"Mean:               {mean_val}")
print(f"Median:             {median_val}")
print(f"Variance (Sample):  {variance_val:.2f}")
print(f"Std Deviation:      {std_dev_val:.2f}")
print("-" * 30)
print(f"SciPy Summary: {description}")
```

```python
# Exercise 3

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Define parameters
mu = 50      # Mean
sigma = 10   # Standard Deviation

# 2. Create a range of x-values (from mu - 4*sigma to mu + 4*sigma)
# This ensures we capture 99.9% of the distribution
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# 3. Calculate the PDF (Probability Density Function)
y = norm.pdf(x, mu, sigma)

# 4. Plotting the results
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'Normal Dist: $\mu={mu}, \sigma={sigma}$', color='royalblue', linewidth=2)
plt.fill_between(x, y, alpha=0.2, color='royalblue') # Fill area for style

# Add labels and title
plt.title('Normal Distribution Curve', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.show()
```

```python
# Exercise 4

import numpy as np
from scipy import stats

# 1. Generate the data
np.random.seed(42) # Ensures you get the same "random" results every time
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# 2. Perform the Independent T-Test
t_stat, p_value = stats.ttest_ind(data1, data2)

# 3. Print the results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value:     {p_value:.4f}")

# 4. Interpret the result
alpha = 0.05
if p_value < alpha:
    print("\nResult: Significant difference found (Reject Null Hypothesis)")
else:
    print("\nResult: No significant difference found (Fail to Reject Null Hypothesis)")
```

```python
# Exercise 5

import numpy as np
from scipy import stats

# 1. Prepare the dataset
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

# 2. Perform Linear Regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

# 3. Predict the price for a 90 sqm house
new_size = 90
predicted_price = intercept + (slope * new_size)

# 4. Results
print(f"Slope (Coefficient): {slope:.2f}")
print(f"Intercept:           {intercept:.2f}")
print(f"R-squared:           {r_value**2:.4f}")
print("-" * 30)
print(f"Predicted price for 90 sqm: {predicted_price:,.2f}")
```

```python
# Exercise 6

import numpy as np
from scipy import stats

# 1. Prepare the dataset
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# 2. Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

# 3. Print the results
print(f"F-value: {f_stat:.4f}")
print(f"P-value: {p_value:.8f}")

# 4. Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nResult: Significant difference found between at least two groups.")
else:
    print("\nResult: No significant difference found.")
```

Answers & Interpretation

1. What are the F-value and P-value?
- F-value: Approximately 28.53.
- P-value: Approximately 0.000027 (or $2.7 \times 10^{-5}$).

2. Based on the P-value, do the fertilizers have significantly different effects on plant growth?
Yes. Since the P-value ($0.000027$) is much smaller than the standard threshold of 0.05, we reject the null hypothesis. This means there is a statistically significant difference in plant growth depending on which fertilizer was used.

3. Explain what would happen if the P-value were greater than 0.05.
If the P-value were greater than 0.05, we would fail to reject the null hypothesis.In practical terms, this would mean that any observed differences in the averages of the three groups were likely due to random chance or natural variation among the plants, rather than the fertilizers themselves being effectively different. You would conclude that all three fertilizers perform roughly the same.
