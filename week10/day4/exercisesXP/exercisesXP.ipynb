```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



# --- Exercise 1: Matrix Operations ---
# Create a 3x3 matrix
matrix = np.array([[6, 1, 1], [4, -2, 5], [2, 8, 7]])
# 1. Calculate determinant
det = np.linalg.det(matrix)
# 2. Find the inverse
inv = np.linalg.inv(matrix)

print("Ex 1 - Matrix Operations:")
print(f"Determinant: {det:.2f}")
print(f"Inverse Matrix:\n{inv}\n")
```

```python
# --- Exercise 2: Statistical Analysis ---
# Generate 50 random numbers
random_array = np.random.rand(50)
print("Ex 2 - Statistics:")
print(f"Mean: {np.mean(random_array):.2f}")
print(f"Median: {np.median(random_array):.2f}")
print(f"Standard Deviation: {np.std(random_array):.2f}\n")
```

```python
# --- Exercise 3: Date Manipulation ---
# Generate dates for January 2023
jan_dates = np.arange('2023-01-01', '2023-02-01', dtype='datetime64[D]')
# Convert to YYYY/MM/DD format (using pandas for formatting ease)
formatted_dates = pd.to_datetime(jan_dates).strftime('%Y/%m/%d')
print("Ex 3 - Dates (First 3):", formatted_dates[:3].tolist(), "\n")
```

```python
# --- Exercise 4: Data Manipulation (NumPy + Pandas) ---
df_rand = pd.DataFrame(np.random.randint(0, 100, size=(5, 3)), columns=['X', 'Y', 'Z'])
# Conditional selection: rows where X > 50
filtered_data = df_rand[df_rand['X'] > 50]
print("Ex 4 - Aggregations:")
print(f"Sum:\n{df_rand.sum()}\nAverage:\n{df_rand.mean()}\n")
```

```python
# --- Exercise 5: Image Representation ---
# Images are arrays of pixels. 0 is black, 255 is white.
grayscale_5x5 = np.random.randint(0, 256, (5, 5), dtype=np.uint8)
print("Ex 5 - 5x5 Grayscale Image Array:\n", grayscale_5x5, "\n")
```

```python
# --- Exercise 6: Basic Hypothesis Testing ---
productivity_before = np.random.normal(loc=50, scale=10, size=30)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)
# Hypothesis: Mean After > Mean Before
improvement = np.mean(productivity_after) - np.mean(productivity_before)
print(f"Ex 6 - Hypothesis Test: Average Productivity Increase is {improvement:.2f}\n")
```

```python
# --- Exercise 7: Complex Array Comparison ---
arr1 = np.array([10, 50, 30])
arr2 = np.array([20, 40, 30])
greater_than = arr1 > arr2
print("Ex 7 - Comparison (is arr1 > arr2?):", greater_than, "\n")
```

```python
# --- Exercise 8: Time Series Slicing ---
time_index = pd.date_range('2023-01-01', periods=365, freq='D')
ts_data = pd.Series(np.random.randn(365), index=time_index)
jan_to_march = ts_data['2023-01':'2023-03']
print(f"Ex 8 - Time Series Slice (Jan-Mar) count: {len(jan_to_march)} days\n")
```

```python
# --- Exercise 9: Data Conversion ---
# NumPy to Pandas
new_df = pd.DataFrame(np.zeros((2, 2)))
# Pandas to NumPy
back_to_np = new_df.to_numpy()
print("Ex 9 - Conversion: Array shape from DF is", back_to_np.shape, "\n")
```

```python
# --- Exercise 10: Basic Visualization ---
plt.figure(figsize=(6, 4))
plt.plot(np.random.randn(20).cumsum(), label='Random Walk', color='red')
plt.title('Ex 10: NumPy Data Visualization')
plt.legend()
plt.show()
```
