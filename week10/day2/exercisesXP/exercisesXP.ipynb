```python
# Exercise 1 : Array Creation and Manipulation

import numpy as np

array = np.arange(10)
print(array)
```

```python
# Exercise 2 : Type Conversion and Array Operations

import numpy as np

lst_float = [3.14, 2.17, 0, 1, 2]
array_float = np.array(lst_float)
array_int = array_float.astype(int)
print(array_int)
```

```python
# Exercise 3 : Working with Multi-Dimensional Arrays

import numpy as np

arr_2d = np.arange(1,10).reshape(3,3)
print(arr_2d)
```

```python
# Exercise 4 : Creating Multi-Dimensional Array with Random Numbers

import numpy as np

arr_rand = np.random.rand(4,5)
print(arr_rand)
```

```python
# Exercise 5 : Indexing Arrays

import numpy as np

array = np.array(
    [[21, 22, 23, 22, 22], [20, 21, 22, 23, 24], [21, 22, 23, 22, 22]])
print(array[1])
```

```python
# Exercise 6 : Reversing elements

import numpy as np

arr = np.arange(10)
arr_inv = arr[::-1]
print(arr_inv)
```

```python
# Exercise 7 : Identity Matrix

import numpy as np

identity = np.identity(4)
print(identity)
```

```python
# 🌟 Exercise 8 : Simple Aggregate Funcs

import numpy as np

arr = np.arange(10)

sum = np.sum(arr)
mean = np.mean(arr)
print(f"Sum: {sum}, Average: {mean}")
```

```python
# 🌟 Exercise 9 : Create Array and Change its Structure

import numpy as np

arr_1d = np.arange(1,21)
arr_2d = arr_1d.reshape(4,5)
print(arr_1d)
print(arr_2d)
```

```python
# 🌟 Exercise 10 : Conditional Selection of Values

import numpy as np

arr = np.arange(10)
arr_even = arr[arr%2 != 0]

print(arr)
print(arr_even)
```
