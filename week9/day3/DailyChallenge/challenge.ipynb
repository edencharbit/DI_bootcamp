```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# 1. Load Data
df = pd.read_csv('USA_cars_datasets.csv')

# 2. Histogram of Prices (Seaborn)
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, color='royalblue')
plt.title('Distribution of Car Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()

# 3. Box Plot (Top 10 Brands)
top_10 = df['brand'].value_counts().head(10).index
df_top_10 = df[df['brand'].isin(top_10)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top_10, x='brand', y='price', palette='viridis')
plt.title('Price Distribution by Brand (Top 10)')
plt.xticks(rotation=45)
plt.show()

# 4. Interactive Scatter Plot (Plotly)
# Run this in your local environment to hover over points!
fig = px.scatter(df, x="mileage", y="price", color="title_status",
                 hover_data=['brand', 'model', 'year'],
                 title="Price vs. Mileage (Interactive)")
fig.show()

# 5. Heatmap (Average Prices)
pivot_table = df_top_10.pivot_table(index='brand', columns='title_status', values='price', aggfunc='mean')
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='RdYlGn')
plt.title('Average Price: Brand vs. Condition')
plt.show()
```
