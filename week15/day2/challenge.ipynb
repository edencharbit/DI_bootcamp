```python
from statsmodels.stats.power import TTestIndPower
import numpy as np

# Initialize the power analysis object
analysis = TTestIndPower()

# Parameters for Task 1
alpha = 0.05
power = 0.8
effect_size_standard = 0.2

# 1. Calculate Required Sample Size for Sweet Bytes
sample_size_bytes = analysis.solve_power(effect_size=effect_size_standard,
                                         alpha=alpha,
                                         power=power,
                                         alternative='two-sided')

print(f"--- Sweet Bytes Checkout Test ---")
print(f"Required sample size per group: {round(sample_size_bytes)}")

# 2. Analyze the Impact of Effect Size (The Chef's Question)
effect_sizes = [0.1, 0.2, 0.3, 0.4]
print(f"\n--- Effect Size Comparison ---")
for es in effect_sizes:
    n = analysis.solve_power(effect_size=es, alpha=alpha, power=power)
    print(f"Effect Size: {es} | Required Sample Size: {round(n)}")
```

Part 2: Impact Analysis
How does the sample size change?
As the effect size gets larger, the required sample size drops sharply.

Why does this happen?
Think of it like tasting sugar in water.

If I add a tiny pinch of sugar (Small Effect Size), you’ll need to take many, many sips (Large Sample Size) to be absolutely sure the water is actually sweeter and you're not just imagining it.

If I dump a whole cup of sugar in (Large Effect Size), you’ll know the water is sweeter after just one or two sips (Small Sample Size).

Mathematically, a larger effect stands out clearly against the "noise" of random customer behavior, so you don't need as much data to prove it's real.

Part 3: The "Sweet Bytes" Team Briefing
How to explain this to the bakers and decorators!

Planning an A/B test is exactly like baking a new recipe. We want to be confident that our new checkout process is better, but we don't want to waste months testing it if we don't have to.

Why the balance matters:

Don't Waste the Dough (Efficiency): If we expect a massive boost in sales (Large Effect), we only need a few hundred customers to prove it. We can finish the test quickly and start making more money sooner!

Avoid 'Ghost' Flavors (Reliability): If we test on too few people, a 'win' might just be a fluke—like one customer who happened to buy 500 cookies by accident. A proper sample size ensures that when we say the new checkout is better, it’s statistically solid, not just a lucky day.

The Goldilocks Zone: By calculating our sample size, we find the 'Just Right' amount of data. We don't test for too long (wasting time) and we don't stop too early (risking a bad decision).
