```python
import pandas as pd
import numpy as np

# Chargement du dataset
df = pd.read_csv('/content/survey.csv')
```

```python
df['treatment_numeric'] = df['treatment'].map({'Yes': 1, 'No': 0})
bins = [18, 25, 35, 45, 55, 75]
labels = ['18-25', '26-35', '36-45', '46-55', '56-75']
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_dist = df.groupby('age_group')['treatment_numeric'].mean() * 100
print("Pourcentage to seek treatment for mental health by age :")
print(age_dist)
```

```python
gender_map = {

    'male': 'Male',
    'm': 'Male',
    'male-ish': 'Male',
    'maile': 'Male',
    'cis male': 'Male',
    'mal': 'Male',
    'male (cis)': 'Male',
    'make': 'Male',
    'man': 'Male',
    'msle': 'Male',
    'mail': 'Male',
    'malr': 'Male',
    'cis man': 'Male',

    'female': 'Female',
    'f': 'Female',
    'trans-female': 'Female',
    'cis female': 'Female',
    'woman': 'Female',
    'femake': 'Female',
    'cis-female/femme': 'Female',
    'female (cis)': 'Female',
    'femail': 'Female'
}

df['Gender_Cleaned'] = df['Gender'].str.lower().str.strip().map(gender_map)

df['Gender_Cleaned'] = df['Gender_Cleaned'].fillna('Other')

print(df['Gender_Cleaned'].value_counts())
```

```python
gender_dist = df.groupby('Gender_Cleaned')['treatment_numeric'].mean() * 100
print("\nPourcentage to seek treatment for mental health by gender :")
print(gender_dist)
```

```python
countries_count = df['Country'].value_counts()
major_countries = countries_count[countries_count >= 10].index
df_major_countries = df[df['Country'].isin(major_countries)]

country_dist = df_major_countries.groupby('Country')['treatment_numeric'].mean() * 100
country_dist = country_dist.sort_values(ascending=False)

print("\nHighest Rates :")
print(country_dist.head(3))

print("\nLowest Rates :")
print(country_dist.tail(3))
```

**Part 1 : Introduction to the Dataset**

1. Mental Health Distribution by Age GroupIn the tech industry, the tendency to seek treatment for mental health conditions shows a clear correlation with age. While younger professionals (18-25) seek treatment less frequently, nearly 71% of employees aged 56-75 have reported seeking professional help. This might suggest that long-term career pressure or a decrease in stigma among senior staff impacts these numbers.

2. Frequency of Mental Health Issues by GenderGender plays a significant role in the frequency of reported mental health issues. Female and non-binary ("Other") employees report seeking treatment at much higher rates than their male counterparts.
- Other/Non-binary: ~82.4% rate of treatment.
- Female: ~68.5% rate of treatment.
- Male: ~45.4% rate of treatment.

3. Geographical Trends (Highest vs Lowest Rates)By filtering for countries with a significant number of respondents (minimum 10), we identified regional differences in mental health reporting:Highest Rates: Australia (61.9%) and the United States (54.7%) lead the reported rates of treatment.Lowest Rates: France (15.4%) and India (30.0%) show the lowest rates, which may be influenced by different cultural stigmas or variations in corporate healthcare benefits.

```python
#Part 2: Data Cleaning

import pandas as pd

df = pd.read_csv('/content/survey.csv')

#Dealing with missing values
missing_counts = df.isnull().sum()
print(df.shape)
missing_pct = (missing_counts / len(df)) * 100
missing_df = pd.DataFrame(
    {
        'Missing Count': missing_counts,
        'Missing %': missing_pct
    })
print(missing_df)

df = df.drop(columns=['comments'])
df.dropna(subset=['self_employed'], inplace=True)
df['work_interfere'] = df['work_interfere'].fillna('Unknown')
df['state'] = df['state'].fillna('Unknown')
print(df.isnull().sum())

# # we keep only the lines with ages from 18 to 75
# df = df[(df['Age'] >= 18) & (df['Age'] <= 75)]

```
