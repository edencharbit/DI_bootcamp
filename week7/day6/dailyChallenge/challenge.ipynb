```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

df = pd.read_csv('/content/datascience_salaries.csv')

# Aperçu des données
print(df.head())
```

```python
scaler = MinMaxScaler()

df['salary_normalized'] = scaler.fit_transform(df[['salary']])

print(df[['salary', 'salary_normalized']].head())
```

```python

df_numeric = df.select_dtypes(include=['number']).dropna()

pca = PCA(n_components=2)
pca_results = pca.fit_transform(df_numeric)

df['pca_1'] = pca_results[:, 0]
df['pca_2'] = pca_results[:, 1]
print(pca.explained_variance_ratio_.sum())
```

```python
salary_stats = df.groupby('experience_level')['salary'].agg(['mean', 'median']).sort_values(by='mean')

salary_stats.columns = ['Average Salary (USD)', 'Median Salary (USD)']

print(salary_stats)
```
