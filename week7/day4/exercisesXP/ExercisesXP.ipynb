```python
"""
==========================================================
EXERCISE 1: IDENTIFYING DATA TYPES
==========================================================

1. A company’s financial reports stored in an Excel file.
   ANSWER: Structured Data
   REASON: The data is organized in a fixed grid of rows and columns with
           defined headers, making it easy to search and analyze.

2. Photographs uploaded to a social media platform.
   ANSWER: Unstructured Data
   REASON: Images consist of raw pixel data and do not follow a
           predefined table or relational model.

3. A collection of news articles on a website.
   ANSWER: Unstructured Data
   REASON: While they may have a date or title, the main content is
           natural language text, which is not organized in a database schema.

4. Inventory data in a relational database.
   ANSWER: Structured Data
   REASON: This is the definition of structured data; it is stored in
           linked tables (SQL) with strict data types and formats.

5. Recorded interviews from a market research study.
   ANSWER: Unstructured Data
   REASON: Audio files are complex data streams that require specialized
           processing (like transcription) to be interpreted by a computer.
"""

```

```python
import pandas as pd
import openpyxl

#Exercise5
data = {
        'Flower_Name': ['Rose', 'Tulip', 'Sunflower', 'Daisy'],
        'Color': ['Red', 'Pink', 'Yellow', 'White'],
        'Quantity': [10, 15, 7, 20]
    }

df = pd.DataFrame(data)
print(df)

excel_filename = "exported_flowers.xlsx"
df.to_excel(excel_filename, sheet_name='Sheet1')
print(f"\nSuccessfully exported to: {excel_filename}")


json_filename = "exported_flowers.json"
df.to_json(json_filename, orient='records')
print(f"Successfully exported to: {json_filename}")
```
