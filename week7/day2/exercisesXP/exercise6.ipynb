```python
import pandas as pd

df_sleep = pd.read_csv('Time Americans Spend Sleeping.csv')

print("First 5 rows of the Sleep Dataset:")
print(df_sleep.head())

print("\nDataset Information:")
df_sleep.info()
```

***A. Year (Trend Analysis)***
Analysis Type: Time-Series / Trend Analysis.

Reasoning: This column allows us to see if the average sleep duration of Americans has increased or decreased over the last decade. By plotting Year on the x-axis and Avg hrs per day sleeping on the y-axis, we can identify long-term societal changes in rest patterns.

***B. Avg hrs per day sleeping (Primary Quantitative Variable)***
Analysis Type: Descriptive Statistics.

Reasoning: This is the core variable of the dataset. It is essential for calculating the mean (average sleep), median, and identifying outliers (people who sleep very little or significantly more than average).

***C. Sex or Age Group (Group Comparison)***
Analysis Type: Comparative Analysis (Categorical).

Reasoning: These columns allow us to ask questions like: "Do women sleep more than men on average?" or "Which age group is the most sleep-deprived?". We can use these to create Grouped Bar Charts or Boxplots to compare distributions across demographics.

***D. Type of Days (Behavioral Analysis)***
