1. How Much Sleep Do Americans Really Get?

Year: **Qualitative (Ordinal)** Even though it's a number, it represents a specific point in time (a label). We don't usually add or multiply years.

Age Group:	**Qualitative**	These are labels for segments of the population (e.g., "18-24").

Sex: **Qualitative**	Categorical data representing gender.

Type of Day: **Qualitative**	Labels for categories like "Weekend" or "Weekday."

Avg hrs per day sleeping:	**Quantitative (Continuous)**	This is a measurement that can be averaged. You can have 7.5 hours or 8.2 hours.


2. Global Trends in Mental Health Disorder

Entity (Country):	**Qualitative**	Names of countries are descriptive labels.
Code	Qualitative	Country codes (e.g., USA, FRA) are unique identifiers, not measurements.

Year:	**Qualitative** (Ordinal)	Represents a chronological sequence.

Schizophrenia (%): 	**Quantitative** (Continuous)	Represents a percentage of the population. It is a measurable ratio.

Depression (%):	**Quantitative** (Continuous)	A numerical value used for statistical comparison and calculation.


3. Credit Card Approvals

A1 (Gender/Status):	**Qualitative**	Binary categories like 'a' or 'b'.

A2 (Age):	**Quantitative** (Continuous)	Represents a numerical age that can be averaged.

A3 (Debt):	**Quantitative** (Continuous)	Financial amounts are measurable and allow for arithmetic operations.

A4 (Bank Customer):	**Qualitative**	Categorical labels indicating if someone is a client or not.

A16 (Approved/Rejected):	**Qualitative**	This is the Target Variable. It is a label representing a decision (+ or -).
